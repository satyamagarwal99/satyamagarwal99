```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head(3)
```

```python
df['zip'].value_counts().head(5)   #top 5 zipcodes for 911 calls
```

```python
df['twp'].value_counts().head(5)
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])   #seperating reaon to make data easily readable
```

```python
df['Reason'].value_counts()            #top reason for calling
```

```python
sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])          #convert timestamp to datetime format
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# It is missing some months! 9,10, and 11 are not there.
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot()             # Could be any column
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

```python
df.groupby('Date').count()['twp'].plot()              #Date column aggregate and a plot of counts of 911 calls
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()                #reason to call is traffic
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()         #reason to call is fire
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()               #reason to call is EMS
plt.title('EMS')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour,cmap='viridis')
```

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
